```python
def countWords(string):
    word_count={}
    word_list= string.split()
    for word in word_list:
        if word in word_count:
            word_count[word]= word_count[word]+1
        else:
            word_count[word]=1
    return word_count
    
```

```python
print (countWords("Это пробный текст текст текст"))
```
